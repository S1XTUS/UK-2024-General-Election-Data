```python
df_winning = df_winning.rename(columns={'constituency_name': 'constituency'})
```

```python
result = df.merge(df_winning[['constituency', 'region_name', 'country_name', 'constituency_type', 'result','ons_id' ]], on='constituency', how='left')
```

```python
new_order = ['ons_id','constituency','region_name','country_name', 'constituency_type',
             'candidate_firstName', 'candidate_surname','party_abbreviation', 'party_name', 'party_displayName', 
             'voteCount', 'votePercentage', 'votePercentageChange','Electorate_size', 'Turnout_perc',
             'new_party', 'old_party','index',  'result' ]
```

```python
result = result.reindex(columns=new_order)
```

```python
result.drop('index', inplace=True, axis=1)
```
